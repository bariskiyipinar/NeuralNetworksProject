```python
import os
import shutil
from sklearn.model_selection import train_test_split

# Ana klasör yolu
images_dir = 'C:/Users/Barış/Desktop/DenemeProje/images/images'  # 'images' klasörünün tam yolu
output_dir = 'C:/Users/Barış/Desktop/DenemeProje'  # Train ve Validation dosyalarını kaydedeceğiniz yer

# Train ve Validation klasörlerini oluşturma
train_dir = os.path.join(output_dir, 'train')
val_dir = os.path.join(output_dir, 'validation')

os.makedirs(train_dir, exist_ok=True)
os.makedirs(val_dir, exist_ok=True)

# Alt klasörlerdeki tüm verileri bulma
all_classes = [d for d in os.listdir(images_dir) if os.path.isdir(os.path.join(images_dir, d))]

# Her sınıfı train ve validation'a bölme
for class_name in all_classes:
    class_dir = os.path.join(images_dir, class_name)
    
    # Resim dosyalarını filtreleyin (örneğin jpg, png)
    images = [f for f in os.listdir(class_dir) if os.path.isfile(os.path.join(class_dir, f)) and f.lower().endswith(('.png', '.jpg', '.jpeg'))]

    if len(images) == 0:
        print(f"Uyarı: {class_name} sınıfında hiç resim yok.")
        continue  # Bu sınıfı atla

    # Eğitim ve doğrulama verilerini ayırma
    train_images, val_images = train_test_split(images, test_size=0.2, random_state=42)

    # Klasörleri oluşturma
    os.makedirs(os.path.join(train_dir, class_name), exist_ok=True)
    os.makedirs(os.path.join(val_dir, class_name), exist_ok=True)

    # Train klasörüne kopyalama
    for img in train_images:
        src = os.path.join(class_dir, img)
        dst = os.path.join(train_dir, class_name, img)
        shutil.copy(src, dst)

    # Validation klasörüne kopyalama
    for img in val_images:
        src = os.path.join(class_dir, img)
        dst = os.path.join(val_dir, class_name, img)
        shutil.copy(src, dst)

    # Yüzdelik oranını yazdırma
    print(f"{class_name} sınıfında {len(images)} resim bulundu.")
    print(f"Train setine {len(train_images)} resim ({len(train_images) / len(images) * 100:.2f}%) eklendi.")
    print(f"Validation setine {len(val_images)} resim ({len(val_images) / len(images) * 100:.2f}%) eklendi.\n")

print("Train ve Validation klasörleri başarıyla oluşturuldu.")
```
